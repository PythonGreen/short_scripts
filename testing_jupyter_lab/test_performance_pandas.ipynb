## Usando "fors" procedurales

```python
month_check = {}
for i in range(2000,2020):
    for j in range(1,13):
        month=str(i)+"{:0>2}".format(j) # o tambien se puede usar rjust(posiciones, relleno) month=str(i)+str(j).rjust(2, '0')        
        #print("Processing: ",month)        
        dt_ini = datetime.datetime.now()        
        df_month = azfd_df["MES_PROCESO"] == month
        
        if True in list(df_month.unique()): #o sino usar empty si estuviese vacio: df_month.empty:        
            azfd_df[df_month].to_csv(r'data/fenix_file_{}.csv'.format(month), index = None, header = True, sep = ';')
            dt_end = datetime.datetime.now()            
            month_check[month] = ['Yes', dt_end - dt_ini ]
            #print("There are data for {}? how does it takes?".format(month), month_check[month])                        
        else:            
            dt_end = datetime.datetime.now()            
            month_check[month] = ['No', dt_end - dt_ini ]
            #print("There are data for {}? how does it takes?".format(month), month_check[month])  
```

```python
print("Total time:", sum([i[1] for i in month_check.values()], datetime.timedelta()), "\n",  
      "Month processed", len(month_check) ,   "\n",  
      "Month with data", len([i for i in month_check.values() if i[0]=='Yes']) ,"\n",  
      "Month with no data", len([i for i in month_check.values() if i[0]=='No'])
     )
```
