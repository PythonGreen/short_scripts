# Using apply

```python
def my_to_csv(partition):
    partition.to_csv(r'data/fenix_file_{}.csv'.format(partition['MES_PROCESO'].values[0]), header = True, sep = ';')
```

```python
%time dd_grouped.apply(my_to_csv, meta=object).compute()
```

# Iterating with loop

MALISIMO en tiempos!!!!

```python
dt_ini = datetime.datetime.now()

elements = dd_grouped.MES_PROCESO.unique()
for group in elements:
    dd_grouped.get_group(group[0]).compute().to_csv(r'data/fenix_file_{}.csv'.format(group[0]), header = True, sep = ';') 
    
dt_end = datetime.datetime.now()    
print(dt_end - dt_ini)
```
